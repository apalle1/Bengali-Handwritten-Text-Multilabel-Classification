### Example 1

With the `Parallel` and `delayed` functions from Joblib, we can simply configure a parallel run of the `my_fun()` function. `n_jobs` is the number of parallel jobs, and we set it to be `2` here. `i` is the input parameter of `my_fun()` function, and we'd like to run 10 iterations.

```
#https://www.youtube.com/watch?v=Ny3O4VpACkc
from joblib import Parallel, delayed
import time, math
```

```
def my_fun(i):
  time.sleep(1)
  return math.sqrt(i**2)

start = time.time()
# n_jobs is the number of parallel jobs
Parallel(n_jobs=2)(delayed(my_fun)(i) for i in range(10))
end = time.time()
print('{:.4f} s'.format(end-start))
```

### Example 2

It’s that easy! What if we have more than one parameters in our functions? It’s also very simple. Let’s define a new function with two parameters `my_fun_2p(i, j)`.

```
def my_fun_2p(i, j):
  time.sleep(1)
  return math.sqrt(i**j)

start = time.time()
# n_jobs is the number of parallel jobs
Parallel(n_jobs=2)(delayed(my_fun_2p)(i, j) for i in range(10) for j in range(3))
end = time.time()
print('{:.4f} s'.format(end-start))
```
